f(x) = -5*x**2 + 10*x - 150

1. Определить корни +
2. Найти интервалы, на которых функция возрастает +
3. Найти интервалы, на которых функция убывает +
4. Построить график +
5. Вычислить вершину
6. Определить промежутки, на котором f > 0 +
7. Определить промежутки, на котором f < 0 +

```python
import math
from sympy import *
from sympy.plotting import plot
```

```python
x = Symbol('x')
fx = -5 * x ** 2 + 10 * x - 150
fx
```

```python
plot(fx) # Строим график
```

```python
sqrt_1, sqrt_2  = solve(fx) # Находим корни
round(sqrt_1.evalf(), 3), round(sqrt_2.evalf())
```

```python
diff(fx) # наодим производную
```

```python
plot(diff(fx))
```

```python
solve(diff(fx)) # Вычислить вершину
```

```python
solve(diff(fx) > 0) # интервалы, на которых функция возрастает
```

```python
solve(diff(fx) < 0) # интервалы, на которых функция возрастает
```

```python
solve(fx > 0) # Определить промежутки, на котором f > 0
```

```python
solve(fx < 0) # Определить промежутки, на котором f < 0
```
